```python
with pd.ExcelWriter('stock_data1.xlsx', engine='xlsxwriter') as writer:
    # Write each dataframe to a different worksheet
    TCS.to_excel(writer, sheet_name='TCS')
    Infosys.to_excel(writer, sheet_name='Infosys')
    HCL.to_excel(writer, sheet_name='HCL')
    LTIMindtree.to_excel(writer, sheet_name='LTIMindtree')
    KPIT.to_excel(writer, sheet_name='KPIT')

print("Data saved to stock_data.xlsx with different worksheets.")
```

```python
# TCS 
TCS = yf.download("TCS.NS", start="2019-01-01", end="2023-12-31")
TCS["TCS_Daily_Returns"] = TCS["Close"].pct_change()
TCS
```

```python
# Calculate price differences
delta = TCS['Close'].diff()

# Set up gains and losses
gains = delta.clip(lower=0)
losses = -1 * delta.clip(upper=0)

# Calculate average gains and losses over a 14-day period (common setting)
avg_gain = gains.ewm(com=14, min_periods=14).mean()
avg_loss = losses.ewm(com=14, min_periods=14).mean()

# Calculate the Relative Strength Index (RSI)
rsi = 100 - (100 / (1 + (avg_gain / avg_loss)))

# Add RSI to the DataFrame
TCS['RSI'] = rsi
```

```python
TCS.head(20)
```

```python
## Computing Volatility
import numpy as np
# Compute the logarithmic returns using the Closing price
TCS['Log_Ret'] = np.log(TCS['Close'] / TCS['Close'].shift(1))

# Compute Volatility using the pandas rolling standard deviation function
TCS['Volatility'] = TCS['Log_Ret'].rolling(window=14).std() * np.sqrt(14)
TCS.head(30)

# Plot the NIFTY Price series and the Volatility
# hdfc[['Close', 'Volatility']].plot(kind = "line", subplots=True, color='blue',figsize=(8, 6))
```

```python
# Calculate the 14-day simple moving average of the closing prices
TCS["Moving_Average"] = TCS['Close'].rolling(window=14).mean()

TCS.head(30)
```

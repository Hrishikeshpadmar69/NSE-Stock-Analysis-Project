```python
# Page_Industries
Page_Ind = yf.download("PAGEIND.NS", start="2019-01-01", end="2023-12-31")
```

```python
Page_Ind["Page_Ind_Daily_Returns"] = Page_Ind["Close"].pct_change()
Page_Ind
```

```python
# Calculate price differences
delta = Page_Ind['Close'].diff()

# Set up gains and losses
gains = delta.clip(lower=0)
losses = -1 * delta.clip(upper=0)

# Calculate average gains and losses over a 14-day period (common setting)
avg_gain = gains.ewm(com=14, min_periods=14).mean()
avg_loss = losses.ewm(com=14, min_periods=14).mean()

# Calculate the Relative Strength Index (RSI)
rsi = 100 - (100 / (1 + (avg_gain / avg_loss)))

# Add RSI to the DataFrame
Page_Ind['RSI'] = rsi
```

```python
Page_Ind.head(20)
```

```python
## Computing Volatility
import numpy as np
# Compute the logarithmic returns using the Closing price
Page_Ind['Log_Ret'] = np.log(Page_Ind['Close'] / Page_Ind['Close'].shift(1))

# Compute Volatility using the pandas rolling standard deviation function
Page_Ind['Volatility'] = Page_Ind['Log_Ret'].rolling(window=14).std() * np.sqrt(14)
Page_Ind.head(30)

# Plot the NIFTY Price series and the Volatility
# hdfc[['Close', 'Volatility']].plot(kind = "line", subplots=True, color='blue',figsize=(8, 6))
```

```python
# Calculate the 14-day simple moving average of the closing prices
Page_Ind["Moving_Average"] = Page_Ind['Close'].rolling(window=14).mean()

Page_Ind.head(30)
```
